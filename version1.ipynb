## Извлечение текста из pdf

```python
from pdf2text import pdf2text
```

## Правила для извлечения логической структуры

```python
import re
RE_HEADER1 = re.compile(r'\d+\D')
RE_HEADER2 = re.compile(r'\d+.\d+\D')
RE_HEADER3 = re.compile(r'\d+.\d+.\d+\D')
RE_LIST = re.compile(r'\d+\)|\w\)|\*|®')
```

```python
def lines2classes(line):
    match = RE_LIST.search(line)
    if match:
        if match.start() == 0:
            return 'list', None
    match = RE_HEADER3.search(line)
    if match:
        if match.start() == 0:
            return 'header', 3
    match = RE_HEADER2.search(line)
    if match:
        if match.start() == 0:
            return 'header', 2
    match = RE_HEADER1.search(line)
    if match:
        if match.start() == 0:
            return 'header', 1
    return 'text', None
```

```python
import json
import os
docs = list(os.walk('docs'))
path = docs[0][0]
docs = docs[0][2]
docs_pred = []
for doc in docs:
    lines = pdf2text(path + '/' + doc)
    res = []
    for line in lines:
        type_, level_ = lines2classes(line)
        elem = {"type": type_, "level": level_, "text": line}
        res.append(elem)
    docs_pred.append(res)
with open("file_predicted.json", "w") as write_file:
    json.dump(docs_pred, write_file, ensure_ascii=False)
docs_pred
```
